```python
def build_query(self, array, index):
        query_body = {
            "query": {
                "constant_score": {
                    "filter": {
                        "bool": {
                            "should": [],
                            "must": []
                        }
                    }
                }
            }
        }
        
        if not array:
            return query_body
        
        if index == TURNAROUND_INDEX:
            fields_search = {"equipmenttag": [], "scopelocation": []}
        else:
            fields_search = {"floc": [], "equipment_id": [], "unit": []}
        
        for item in array:
            val = item["value"].strip()
            wildcard = "*" in val

            if not wildcard:
                wildcard_value = re.sub(r'[^0-9a-zA-Z_äöüßÄÖÜ]+', '*', val)
            else:
                for k in fields_search.keys():
                    fields_search[k].append((val, wildcard))
                continue
            
            q = {"should": [], "must": []}
            if item["tag"].lower() == "all":
                for k in fields_search.keys():
                    fields_search[k].append((wildcard_value, wildcard))
            else:
                q["must"] += self.build_wildcard_query(item["tag"].lower(), f"*{wildcard_value}*")
            
            if q["should"]:
                query_body["query"]["constant_score"]["filter"]["bool"]["should"] += q["should"]
            if q["must"]:
                query_body["query"]["constant_score"]["filter"]["bool"]["must"] += q["must"]
        
        should_s = []
        for field, items in fields_search.items():
            if not items:
                continue
            queries = [x[0] for x in items]
            wildcard = any(x[1] for x in items)

            wildcard_value = '*'.join(queries)
            if not wildcard:
                wildcard_value = "*" + wildcard_value + "*"
            should_s += self.build_wildcard_query(field, wildcard_value)
        
        if should_s:
            query_body["query"]["constant_score"]["filter"]["bool"]["must"].append({"bool": {"should": should_s}})
        
        return query_body
```

```python
import jwt

from fastapi.security import HTTPBearer

# from app.core.config import DOMAIN, API_AUDIENCE, ALGORITHMS, ISSUER
DOMAIN="shell-sep-dev.eu.auth0.com"
API_AUDIENCE="https://sep-backend.coacapp.de"
ALGORITHMS="RS256"
ISSUER="https://shell-sep-dev.eu.auth0.com/"
token_auth_scheme = HTTPBearer()

class VerifyToken():
    """Does all the token verification using PyJWT"""

    def __init__(self, token):
        self.token = token
        self.config = {
          "DOMAIN": DOMAIN,
          "API_AUDIENCE": API_AUDIENCE,
          "ALGORITHMS": ALGORITHMS,
          "ISSUER": ISSUER
        }
        # This gets the JWKS from a given URL and does processing so you can
        # use any of the keys available
        jwks_url = f'https://{self.config["DOMAIN"]}/.well-known/jwks.json'
        print(jwks_url)
        self.jwks_client = jwt.PyJWKClient(jwks_url)

    def verify(self):
        # This gets the 'kid' from the passed token
        try:
            self.signing_key = self.jwks_client.get_signing_key_from_jwt(
                self.token
            ).key
        except jwt.exceptions.PyJWKClientError as error:
            print("here")
            return {"status": "error", "msg": error.__str__()}
        except jwt.exceptions.DecodeError as error:
            print("here2")
            return {"status": "error", "msg": error.__str__()}

        try:
            payload = jwt.decode(
                self.token,
                self.signing_key,
                algorithms=self.config["ALGORITHMS"],
                audience=self.config["API_AUDIENCE"],
                issuer=self.config["ISSUER"],
            )
        except Exception as e:
            print("here3")
            return {"status": "error", "message": str(e)}

        return payload
```

```python
import pandas as pd
import re
import ast

from app.models.core import FileType
from app.utils.utility import WIMS_MATCHES_MAPPING, TUEV_MATCHES_MAPPING
from app.db.models import equipments, inspections, tuev_matches
from app.db.tasks import preprocess_df_before_upload, preprocess_equipments_df_before_db_upload, preprocess_inspections_df_before_db_upload, upsert_to_db

from app.services.data_transformations import DataTransformations
from app.services.filter_logic import FilterLogic
from app.services.appointment_logic import AppointmentLogic

from collections import Counter


class Matching(): 
    def __init__(self, df_shell, df_tuev, df_specifications): 
        self.filter_logic = FilterLogic(df_specifications)
        self.data_transformations = DataTransformations()
        self.df_shell = df_shell
        self.df_shell["WKP"] = self.df_shell.apply(lambda row: self.data_transformations.wkp_manip(row), axis=1)
        self.df_shell["FLOC_clean"] = self.df_shell.apply(lambda row: self.data_transformations.clean_location_from_FLOC(row), axis = 1)
        self.df_shell["FLOC_zero_clean"] = self.df_shell.apply(lambda row: self.data_transformations.find_strings_to_clean(row), axis = 1)
        try:
            df_specifications = df_specifications.rename(columns = {r"Equipment\Name":"EquipmentID"})
        except:
            pass                                         
                                     
        self.df_shell = pd.merge(self.df_shell, df_specifications, how = "left", left_on="Equipment ID", right_on="EquipmentID")
        self.df_tuev = df_tuev
        self.df_tuev = self.df_tuev.apply(lambda row: self.data_transformations.remove_udr_idr(row), axis = 1)
    
    def equipment_matching(self, df_shell, df_tuev): 
        equipment_matches = []
        df_tuev["Inventar_clean"] = df_tuev.apply(lambda row: str(row["Inventar-Nr."]).replace("-", " "), axis = 1)
        df_tuev["Herstell_clean"] = df_tuev.apply(lambda row: str(row["Herstell-Serial-Nr."]).replace("-", " "), axis = 1)
        for _, row in df_shell.iterrows(): 
            floc = row["FLOC"]
            floc_clean = row["FLOC_clean"]
            floc_stripped = row["FLOC"].replace("-", " ")
            floc_zero_clean = row["FLOC_zero_clean"]
            equi_desc = re.findall(r"\w{1,3}-\d{4,5}-\w{1,2}", row["Equipment Description"])
            equi_desc = equi_desc[0] if equi_desc else False
            
            matched_equipments_tuev = list()
            for item in [floc, floc_clean, floc_stripped, floc_zero_clean, equi_desc]: 
                matched_equipments_tuev.extend(self.query_tuev_data(item, df_tuev))
            print("----------------------------------------------------------------")
            print(len(matched_equipments_tuev))
            print("----------------------------------------------------------------")
            
            if len(matched_equipments_tuev) > 0: 
                equipment_matches.append({"Shell Equipment ID": row["Equipment ID"], "TUEV Equipment ID": list(set(matched_equipments_tuev)), "Shell Terminplan": row["Terminplan"], "match_label": True})
        df_matches = pd.DataFrame.from_records(equipment_matches)
        df_shell_w_matches = pd.merge(df_shell, df_matches, how="left", left_on=["Equipment ID", "Terminplan"], right_on=["Shell Equipment ID", "Shell Terminplan"])
        return df_shell_w_matches
    
    def appointment_check(self, df_shell, df_tuev): 
        appointment_logic = AppointmentLogic()
        df_matches = df_shell[df_shell["match_label"]==True]
        df_unmatches = df_shell[pd.isnull(df_shell["match_label"])]
        
        df_matches = df_matches.apply(lambda row: self.iterate_appointments(row, df_tuev, appointment_logic), axis = 1)
        return pd.concat([df_matches, df_unmatches])
        #return df_matches, df_unmatches
                
    
    
    def iterate_appointments(self, row, df_tuev, appointment_logic):
        tuev_items = row["TUEV Equipment ID"]
        #tuev_items = ast.literal_eval(row["TUEV Equipment ID"])
        appointment_list = list()
        a = 0
        b = 0
        for item in tuev_items: 
            res = df_tuev.query("`Equipment-Nr.` == @item")
            for __, tuev_row in res.iterrows(): 
                tmp_row = row.append(tuev_row)
                appointment_val = appointment_logic.run_schedule_matching(tmp_row)
                if isinstance(appointment_val, bool):
                    appointment_list.append((appointment_val, tmp_row))
                    if appointment_val:
                        a+=1
                    else: 
                        b+=1
        return self.appointment_selector(appointment_list, row)
    
    def appointment_selector(self, appointment_list, row):
        z =Counter([item[0] for item in appointment_list])
        if z[True] >= 1: 
            for item in appointment_list:
                if item[0] == True: 
                    row = item[1]
                    row["appointment_label"] = True
                    #if multiple True eg IU/DP then only one is returned
                    
        elif z[False] >= 1: 
            for item in appointment_list:
                if item[0] == False: 
                    row = item[1]
                    row["appointment_label"] = False
                    
        else: 
            row["appointment_label"] = "No matching appointment available"
        
        return row
                
                     
    def query_tuev_data(self, shell_data_item, df_tuev): 
        query_res_inv = df_tuev.query("`Inventar-Nr.` == @shell_data_item")
        query_res_inv_clean = df_tuev.query("Inventar_clean == @shell_data_item")
        query_res_herstell = df_tuev.query("`Herstell-Serial-Nr.` == @shell_data_item")
        query_res_herstell_clean = df_tuev.query("Herstell_clean == @shell_data_item")
        res_df = pd.concat([query_res_inv, query_res_inv_clean, query_res_herstell, query_res_herstell_clean]).drop_duplicates(["Equipment-Nr."])

        return list(res_df["Equipment-Nr."]) if not res_df.empty else []
    
    def main(self): 
        df_shell_date_valid = self.df_shell[(self.df_shell["Fälligkeit"] >= str(self.df_tuev.NPD.min())) & (self.df_shell["Fälligkeit"] <= str(self.df_tuev.NPD.max()))]
        df_shell_date_invalid = self.df_shell[self.df_shell["Fälligkeit"] >= str(self.df_tuev.NPD.max())]
        df_shell_date_invalid["filter_label"] = False
        df_shell_true_filter, df_shell_false_filter = self.filter_logic.create_filtered_dataset(df_shell_date_valid)
        equipment_matches = self.equipment_matching(df_shell_true_filter, self.df_tuev)
        appointment_matches = self.appointment_check(pd.concat([equipment_matches, df_shell_false_filter,df_shell_date_invalid]), self.df_tuev)
        
        tuev_df = appointment_matches[appointment_matches["match_label"]==True]
        join_cols = ["Equipment-Nr.", "Prüf-Ident", "NPD", "Inventar-Nr.", "Herstell-Serial-Nr."]
        tuev_matches = self.df_tuev.merge(tuev_df[join_cols + ["match_label", "Equipment ID"]], how="left", left_on=join_cols, right_on=join_cols)
        tuev_matches = tuev_matches[self.df_tuev.columns.to_list() + ["match_label", "Equipment ID"]]
        tuev_matches = tuev_matches.rename(columns={"match_label": "matched", "Equipment ID": "shell_equipment_id"})
        return appointment_matches, tuev_matches
    

def matching_algo(wims_raw_df, tuev_raw_df, specifications_df):
    model = Matching(wims_raw_df, tuev_raw_df, specifications_df)
    appointment_matches, tuev_matches_df = model.main()

    # INSPECTIONS and EQUIPMENTS data
    wims_matches_df = preprocess_df_before_upload(appointment_matches.copy(), FileType.wims_matches)
    wims_matches_df = wims_matches_df.rename(columns=WIMS_MATCHES_MAPPING)

    equipments_df = preprocess_equipments_df_before_db_upload(wims_matches_df)
    inspections_df = preprocess_inspections_df_before_db_upload(wims_matches_df)

    upsert_to_db(equipments_df, equipments, FileType.wims_matches)
    upsert_to_db(inspections_df, inspections, FileType.wims_matches, manual_match=True)

    # TUEV_MATCHES data
    tuev_matches_df = preprocess_df_before_upload(tuev_matches_df, FileType.tuev_matches)
    ## Rename excel cols to db fields
    tuev_matches_df = tuev_matches_df.rename(columns=TUEV_MATCHES_MAPPING)
    upsert_to_db(tuev_matches_df, tuev_matches, FileType.tuev_matches)
```

```python
class FilterLogic(): 
    def __init__(self, specifications_df): 
        specifications_df["label"] = specifications_df.apply(lambda row: self.check_specification_apply(row), axis = 1)
        self.match_df = specifications_df[specifications_df['label'] == True]
        
    
    def check_specification_apply(self, row): 
        selbst_rohr = False
        wgk = False
        pruefgrundlage = False
        #pruefgrundlage = True


        if isinstance(row["SELBST. ROHRLEITUNGSANLAGE"], str) and row["SELBST. ROHRLEITUNGSANLAGE"].lower() == "ja": 
            selbst_rohr = True
        if isinstance(row["WGK (WASSERGEFÄHRDUNGSKLASSE)"], str) and row["WGK (WASSERGEFÄHRDUNGSKLASSE)"].lower() in ["wgk 3", "wgk 1", "wgk 2"]: 
            wgk = True
        if isinstance(row["EQUIPMENT ID PRÜFGRUNDLAGE"], str) and row["EQUIPMENT ID PRÜFGRUNDLAGE"] not in ["NICHT WHG RELEVANT", "NICHT WIEDERKEHREND PRÜFPFLICHTIG (GEFÄHRDUNGSSTUFE A ODER B)"]: 
            pruefgrundlage = True

        return all([selbst_rohr, wgk, pruefgrundlage])
    
    def is_ko_list(self, row): 
        ko_list = ["M_RLF_", "M_VLV_NRET", "M_PIP_SLR", "M_PIP_DUCT", "M_PIP_SYS"]
        sap_type = row['SAP Types']
        if not isinstance(sap_type, str): 
            return False
        if any(item in row['SAP Types'] for item in ko_list): 
            return False
        return True

    def is_rohrleitung(self, row): 
        if row["Equipment Group"].lower() == "rohrleitung":
            return True
        return False

    def is_SV(self, row): 
        description = row['Description'].split()
        for item in description: 
            if "SV" in item and item != "BSV":
                return True
        return False

    def is_sap_type(self, row): 
        yes = ["M_PIP_X", "M_PIP_OFF", "M_PIP_PROC", "M_PIP_HAZA", "M_PIP_FLOA"]
        no = ["M_PIP_FLEX", "M_PIP_FLAR"]

        if any(item in row['SAP Types'] for item in yes): 
            return True
        elif any(item in row['SAP Types'] for item in no): 
            return False
        else: 
            return "UNKNOWN"

    def is_awsv(self, row): 
        if not "IU" in row["Description"]:
            return "awsv" in row["Description"].lower()
        else: 
            return False

    def is_interval(self, row): 
        interv = row["Exact Time Interval"]
        return interv == 60.0

    def check_recurring(self, row): 
        if pd.isnull(row["WKP"]):
            return False
        else: 
            return row["WKP"]

    def create_label_by_decision_tree(self, row):
        if self.check_recurring(row): 
            #1. Bestandteil KO Liste?
            if not self.is_ko_list(row):
                #1. Ja
                return False
            #1. Nein
            #2. Equipmentgruppe = Rohrleitung?
            elif self.is_rohrleitung(row):
                #2. Ja
                #2.1 SAP_Typ = M_PIP
                if self.is_sap_type(row) == "UNKNOWN":
                    #2.1 not in yes or no
                    return "UNKNOWN"
                elif self.is_sap_type(row):
                    if self.is_awsv(row):
                        if not self.is_interval(row):
                            return False
                        #if not is_interval(row):
                    #2.1 ja
                    #2.2 spezifikation
                        if row["Equipment ID"] in list(self.match_df["EquipmentID"]):
                            #2.2 ja
                            return True
                        else:
                            #2.2 nein
                            return False
                    else:
                        return False
            #2.1 nein
            #2. Nein
            elif self.is_SV(row): 
                return True
            else:
                return False
        else: 
            #print(row["WKP"])
            description = row["Description"]
            if ("AOP" in description or "Nachprüfung" in description or " NP " in description): 
                return True
            else: 
                return False

    def create_filtered_dataset(self, shell_df): 
        shell_df['filter_label'] = shell_df.apply(lambda row: self.create_label_by_decision_tree(row), axis = 1)
        return shell_df[shell_df["filter_label"]], shell_df[~shell_df["filter_label"]]
```
